```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'C:\Users\gowth\OneDrive\Desktop\Vserve Assessment\Assessment_analyst\assessment_melb.xlsx')
print(df.shape)
df.head()
```

```python
#Checking Null Values
print("Null Values in Each Column:")
print(df.isnull().sum())
```

```python
#Treat Null Values
# Numeric columns - fill with median
num_cols = ['Price', 'Bedroom2', 'Bathroom', 'Car',
            'Landsize', 'BuildingArea', 'YearBuilt']

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns - fill with mode
cat_cols = ['CouncilArea', 'Regionname']

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("After Null Treatment:")
print(df.isnull().sum())
```

```python
#Detect & Treat Outliers
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

before = df.shape[0]

df = df[
    (df['Price'] >= Q1 - 1.5 * IQR) &
    (df['Price'] <= Q3 + 1.5 * IQR)
]

after = df.shape[0]
print(f"Outliers Removed : {before - after}")
print(f"Remaining Rows   : {after}")
```

```python
# RATIONALE:
# 1. Numeric Nulls → Median
#    Reason: Median is not affected by extreme values

# 2. Categorical Nulls → Mode
#    Reason: Most frequent value is the best replacement

# 3. Outliers → IQR Method on Price
#    Reason: Property prices have extreme values
#    that can skew the entire analysis
```

```python
#Task 2: Feature Engineering (Property-Level)
#     Compute the following derived metrics for each property/transaction
```

```python
# Price per Room
df['Price_per_Room'] = df['Price'] / df['Rooms'].replace(0, np.nan)

# Price per Land Square Meter
df['Price_per_Sqm'] = df['Price'] / df['Landsize'].replace(0, np.nan)

# Property Age (Current Year = 2017)
df['Property_Age'] = 2017 - df['YearBuilt']

print(df[['Suburb', 'Price', 'Price_per_Room',
          'Price_per_Sqm', 'Property_Age']].head(10))
```

```python
#Task 3: Feature Engineering (Suburb-Level)
#     Compute the following aggregated metric for each suburb:
#     Ratio of Average Price to Average Building Area
```

```python
# Ratio of Average Price to Average Building Area
suburb_stats = df.groupby('Suburb').agg(
    Avg_Price        = ('Price', 'mean'),
    Avg_BuildingArea = ('BuildingArea', 'mean')
).reset_index()

# Avoid division by zero
suburb_stats['Price_to_BuildingArea_Ratio'] = (
    suburb_stats['Avg_Price'] /
    suburb_stats['Avg_BuildingArea'].replace(0, np.nan)
)

print(suburb_stats.head(10))
```
